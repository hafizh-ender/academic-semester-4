```python
import pandas as pd
import matplotlib.pyplot as plt

fivehz = pd.read_csv('5hz.csv')
sixhz = pd.read_csv('6hz.csv')
sevenhz = pd.read_csv('7hz.csv')

fivehz
```

```python
xfive = fivehz['X--Trace 1::[CH1]']
yfive = fivehz['Y--Trace 1::[CH1]']

plt.plot(xfive, yfive)

plt.xlim(-1, 1)
plt.ylim(-3.5, 3.5)

plt.xlabel("Time (200 ms)")
plt.ylabel("Voltage (V)")
```
